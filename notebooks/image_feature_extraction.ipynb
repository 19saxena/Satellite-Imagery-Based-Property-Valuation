```python
import pandas as pd
train=pd.read_excel("../data/raw/train.xlsx")
test=pd.read_excel("../data/raw/test2.xlsx")
train=train.drop(columns=["date"])
test=test.drop(columns=["date"])
```

```python
def attach_images(df):
    df=df.copy()
    df["image_path"]=df["id"].astype(str).apply(
        lambda x:f"../data/images/all/{x}.png"
    )
    return df
train=attach_images(train)
test=attach_images(test)
```

```python
extract(train,"../data/images/all", "../outputs/train_img_features.csv")
extract(test,"../data/images/all", "../outputs/test_img_features.csv")
```
